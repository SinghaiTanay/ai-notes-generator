```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("groq:llama-3.1-8b-instant")
```

```python
from pydantic import BaseModel, Field

class notes(BaseModel):
    topic: str = Field(description="Name of the topic")
    definition: str = Field(description="Definition of the topic")
    important_points: list[str] = Field(description="important points of the topic")
    advantages: list[str] = Field(description="advantage of the topic")
    disadvantages: list[str] = Field(description="disadvantage of the topic")
    exam_tip: str = Field(description="Exam tip")
```

```python
str_model = model.with_structured_output(notes)
```

```python
topic = input("Enter Topic Name:")

prompt = f"""
You are an expert teacher.

Explain {topic} in simple language for a B.Tech student.

Provide:
- Topic Name
- Definition
- 5 Important Points
- Advantages
- Disadvantages
- Exam Tip

Keep the explanation concise and exam-oriented.
"""


response = str_model.invoke(prompt)
```

```python
response
```

```python
print(f"Topic: {response.topic}")
print(f"\nDefinition:\n{response.definition}")

print("\nImportant Points:")
for point in response.important_points:
    print(f"- {point}")

print("\nAdvantages:")
for adv in response.advantages:
    print(f"- {adv}")

print("\nDisadvantages:")
for dis in response.disadvantages:
    print(f"- {dis}")

print(f"\nExam Tip:\n{response.exam_tip}")
```
